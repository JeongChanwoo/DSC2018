```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import os
```

```python
file_path = './datasets/credit'
```

```python
index_file = os.path.join(file_path, 'index.txt')
```

```python
column_name = ['Status of existing checking account',
              'Duration in month',
              'Credit history',
              'Purpose',
              'Credit amount',
               'Savings account/bonds',
              'Present employment since',
              'Installment rate in percentage of disposable income',
              'Personal status and sex',
              'Other debtors / guarantors',
              'Present residence since',
              'Property',
              'Age in years',
              'Other installment plans',
              'Housing',
              'Number of existing credits at this bank',
              'Job',
              'Number of people being liable to provide maintenance for',
              'Telephone',
              'foreign worker',
              'Target_value']
```

```python
data_file = pd.read_csv(os.path.join(file_path, 'german.data.txt'),sep=' ', names=column_name)
```

```python
data_file.index = np.arange(1,1001)
```

```python
col = data_file.columns
print(col)
```

```python
data_id = data_file.index.tolist()
```

```python
y = data_file.Target_value
x = data_file.drop('Target_value', axis = 1)
x.head()
```

```python
numeric_type = ['Duration in month', 
                'Credit amount',
                'Age in years', 
                'Installment rate in percentage of disposable income',
                'Present residence since',
                'Number of existing credits at this bank',
                'Number of people being liable to provide maintenance for',
               ]

category_type = [category for category in col.tolist()[:-1] if category not in numeric_type]

certain_numeric_type = ['Duration in month', 'Credit amount', 'Age in years']
cat_numeric_type = [ column for column in numeric_type if column not in certain_numeric_type]
```

```python
from sklearn.preprocessing import LabelBinarizer
```

```python
lb = LabelBinarizer()
y_data = lb.fit_transform(y)
y_data[:10]
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
x_data = x.copy()
```

```python
label_column = category_type
label_encoder_list = []

for column_name in label_column:
    le = LabelEncoder()
    le.fit(x_data[column_name])
    x_data[column_name] = le.transform(x_data[column_name])
    label_encoder_list.append(le)
    
    del le
```

```python
x_data.head()
```

```python
x_data_stack = x_data.copy()
x_data_stack.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import SGDClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import cross_val_predict
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
```

```python
from sklearn.metrics import confusion_matrix
```

```python
kfold = KFold(n_splits=10 , random_state=22)
average_cv = []
accuracy = []
std_cv=[]

classifiers = ['Logistic_Reg', 'SGD_classfier', 'KNN', 'Naive_bayes', 'SVM','Decision_tree', 'Random_forest']
models = [LogisticRegression(), SGDClassifier(), KNeighborsClassifier(n_neighbors=9),GaussianNB(),
          svm.SVC(kernel='rbf'),DecisionTreeClassifier(),RandomForestClassifier(n_estimators=100)]

for model in models:
    
    cv_result = cross_val_score(model,x_data_stack, y_data, cv = kfold, scoring='accuracy')
    average_cv.append(cv_result.mean())
    std_cv.append(cv_result.std())
    accuracy.append(cv_result)

baseline_estimate_df = pd.DataFrame({'cv_mean' : average_cv,
                                                        'cv_std' : std_cv},index = classifiers)
baseline_estimate_df
```

```python
plt.figure(figsize=(12,6))
plt.grid()
plt.yticks(np.arange(0,1,0.05))

sns.boxplot(x=classifiers, y= accuracy)
```

```python
plt.figure(figsize=(10,5))
plt.title('Average cv accuracy')
sns.barplot(x=baseline_estimate_df['cv_mean'], y = classifiers)
```

```python
figure,ax =plt.subplots(3,3, figsize=(12,12))
y_pred = cross_val_predict(LogisticRegression(),x_data_stack, y_data,cv=5)
sns.heatmap(confusion_matrix(y_data, y_pred),ax=ax[0,0],annot=True,fmt='2.0f')
ax[0,0].set_title('Matrix of Logisitc')
y_pred = cross_val_predict(SGDClassifier(),x_data_stack, y_data,cv=5)
sns.heatmap(confusion_matrix(y_data,y_pred),ax=ax[0,1],annot=True,fmt='2.0f')
ax[0,1].set_title('Matrix for SGD')
y_pred = cross_val_predict(KNeighborsClassifier(n_neighbors=10),x_data_stack, y_data,cv=5)
sns.heatmap(confusion_matrix(y_data, y_pred),ax=ax[0,2],annot=True,fmt='2.0f')
ax[0,2].set_title('Matrix for KNN')
y_pred = cross_val_predict(GaussianNB(),x_data_stack, y_data,cv=5)
sns.heatmap(confusion_matrix(y_data, y_pred),ax=ax[1,0],annot=True,fmt='2.0f')
ax[1,0].set_title('Matrix for Gaussian')
y_pred = cross_val_predict(svm.SVC(kernel='rbf'),x_data_stack, y_data,cv=5)
sns.heatmap(confusion_matrix(y_data, y_pred),ax=ax[1,1],annot=True,fmt='2.0f')
ax[1,1].set_title('Matrix for SVM_rbf')
y_pred = cross_val_predict(DecisionTreeClassifier(),x_data_stack, y_data,cv=5)
sns.heatmap(confusion_matrix(y_data, y_pred),ax=ax[1,2],annot=True,fmt='2.0f')
ax[1,2].set_title('Matrix for Decision Tree')
y_pred = cross_val_predict(RandomForestClassifier(n_estimators=100),x_data_stack, y_data,cv=5)
sns.heatmap(confusion_matrix(y_data, y_pred),ax=ax[2,0],annot=True,fmt='2.0f')
ax[2,0].set_title('Matrix for RandomForest')
```

```python
plt.figure(figsize=(12,12))
plt.title('Feature Importance in Random Forests')
model=RandomForestClassifier(n_estimators=500,random_state=0)
model.fit(x_data_stack,y_data)
pd.Series(model.feature_importances_,x_data.columns).sort_values(ascending=True).plot.barh(width=0.8)
```
