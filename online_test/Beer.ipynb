```python
df_recipe3 = df_recipe2.copy()
```

```python
def PlatoToSG(Plato):
    #Change Plato Sugar Scale to Specific Gravity Sugar Scale
    return 1+(Plato/(258.6-((Plato/258.2)*227.1)))
```

```python
for sugar in range(len(df_recipe3)):
    if df_recipe3.loc[sugar, 'SugarScale']=='Plato':
        df_recipe3.loc[sugar,'SugarScale']= 'Specific Gravity'
        df_recipe3.loc[sugar,'OG']= PlatoToSG( df_recipe3.loc[sugar,'OG'])
        df_recipe3.loc[sugar,'FG'] = PlatoToSG( df_recipe3.loc[sugar,'FG'])
        df_recipe3.loc[sugar,'BoilGravity'] = PlatoToSG(df_recipe3.loc[sugar,'BoilGravity'])
```

```python
df_recipe3.groupby('SugarScale').mean()[['OG','FG','BoilGravity']]
```

```python
# Now all the scales in the value is Speicific Gravity, thus drop SugarScale
df_recipe3 = df_recipe3.drop('SugarScale',axis=1)
```

```python
Boil_Gravity_Drop2 =df_recipe3['BoilGravity'].dropna()

fig=plt.figure(figsize=(20,10))
sns.distplot(Boil_Gravity_Drop2)
```

```python
Boil_Gravity_Drop2.describe()
```

```python
#The distribution is rightward skewed due to the outlier. Thus it is better to set Null as median value
df_recipe3['BoilGravity'] = df_recipe3['BoilGravity'].fillna(df_recipe3['BoilGravity'].median())
```
