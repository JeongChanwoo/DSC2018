```python
def PlatoToSG(Plato):
    #Change Plato Sugar Scale to Specific Gravity Sugar Scale
    return 1+(Plato/(258.6-((Plato/258.2)*227.1)))
```

```python
for sugar in range(len(df_recipe3)):
    if df_recipe3.loc[sugar, 'SugarScale']=='Plato':
        df_recipe3.loc[sugar,'SugarScale']= 'Specific Gravity'
        df_recipe3.loc[sugar,'OG']= PlatoToSG( df_recipe3.loc[sugar,'OG'])
        df_recipe3.loc[sugar,'FG'] = PlatoToSG( df_recipe3.loc[sugar,'FG'])
        df_recipe3.loc[sugar,'BoilGravity'] = PlatoToSG(df_recipe3.loc[sugar,'BoilGravity'])
```

```python
df_recipe3.groupby('SugarScale').mean()[['OG','FG','BoilGravity']]
```

```python
# Now all the scales in the value is Speicific Gravity, thus drop SugarScale
df_recipe3 = df_recipe3.drop('SugarScale',axis=1)
```

```python
# Size(L) ~= Boilsize (0.99)
# OG ~= ABV (0.97)
# OG ~=FG (0.79)
# FG ~=ABV (0.64)

# One can quickly see that there seems to be a near perfect correlation between “Size.L.” and “BoilSize”, 
# which appears to be redundant information, so I will go ahead and drop the “Size.L.” feature. 

# Boilsize와 Size(L)가 .99의 상관을 보이는 이유는 Boilsize가 끌이기 전의 액체의 양을, 
# Size(L)는 완성 후의 맥주의 양을 나타내기 때문이다.
# Recipe를 이용해서 Beer Style 추측하는 것이기 떄문에 Size(L)보다는 Boilisze를 남기는 것이 더 적절 할 것이다.


#Additionally, there seems to be a strong correlation between “OG” and “ABV” and a semi-strong one between “OG” and “FG”. 
# This may be an issue with collinearity or potential multi-collinearity, So let’s investigate…
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import cross_val_predict
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.metrics import confusion_matrix
```

```python
classifiers = ['SGD_classfier', 'KNN', 'Naive_bayes', 'SVM','Decision_tree', 'Random_forest']
print(classifiers[0], 'is done')
```

```python
kfold = KFold(n_splits=5 , random_state=777)
average_cv = []
accuracy = []
std_cv=[]

classifiers = ['SGD_classfier', 'KNN', 'Naive_bayes', 'SVM','Decision_tree', 'Random_forest']
models = [SGDClassifier(), KNeighborsClassifier(n_neighbors=9),GaussianNB(),
          svm.SVC(kernel='rbf'),DecisionTreeClassifier(),RandomForestClassifier(n_estimators=100)]

start_time = time.time()
i=0
for model in models:
    
    cv_result = cross_val_score(model,X_data, Y_data, cv = kfold, scoring='accuracy')
    average_cv.append(cv_result.mean())
    std_cv.append(cv_result.std())
    accuracy.append(cv_result)
    print(classifiers[i], 'is done')
    print("----%s senconds----"%(time.time() - start_time))
    i = i +1
    
baseline_estimate_df = pd.DataFrame({'cv_mean' : average_cv,
                                                        'cv_std' : std_cv},index = classifiers)
baseline_estimate_df
```

```python
baseline_estimate_df.index=['SGD Classfier', 'KNN', 'Naive Bayes', 'SVM','Decision Tree', 'Random Forest']
```
