```python
df_recipe2['SugarScale'].value_counts()
```

```python
def PlatoToSG(Plato):
    return 1+(Plato/(258.6-((Plato/258.2)*227.1)))
```

```python
df_recipe3 = df_recipe2.copy()
for sugar in range(len(df_recipe3)):
    if df_recipe3['SugarScale'][sugar]=='Plato':
        df_recipe3['SugarScale'][sugar] = 'Specific Gravity'
        df_recipe3['OG'][sugar] = PlatoToSG(df_recipe3['OG'][sugar])
        df_recipe3['FG'][sugar] = PlatoToSG(df_recipe3['FG'][sugar])
        df_recipe3['BoilGravity'][sugar] = PlatoToSG(df_recipe3['BoilGravity'][sugar])
```
